```python
import pandas as pd

# 뉴스 파일 로드

data=pd.read_excel('C:\\Users\\user\\Desktop\\기타\\online_news_pos.xlsx')
```

```python
# 확인

data['pos_t'][1]
data['pos_c'][1]
```

# title 불용어 제거를 위한 전처리

```python
# pos_t 리스트안의 문자형 분리 & 특수문자 제거

data_sp=[]
for i in range(len(data)):
    data_sp.append(data['pos_t'][i].split(','))
    
data_sp[0]
```

```python
data_st=[]
for i in range(len(data_sp)):
    data_st2=[]
    for j in data_sp[i]:
        data_st2.append(j.lstrip('[').rstrip('"]').strip(" ' "))
    data_st.append(data_st2)

data_st[0]
```

```python
# 불용어 로드

stop_word=pd.read_excel('C:\\Users\\user\\Desktop\\기타\\stopwords_v0.87.xlsx')
```

```python
# 불용어를 리스트로 변환

SW=[]
for k in stop_word['단어']:
    SW.append(k)
```

```python
SW
```

```python
# 불용어 리스트 set으로 변환

a=set(SW)
```

```python
a
```

# title 불용어 제거

```python
# 불용어 제거

SW_del=[]
for i in data_st:
    b=set(i)
    SW_del.append(b-a)
```

```python
# 확인

SW_del[:2]
```

```python
# set 변환이 된 데이터를 다시 리스트로 변환

SW_del2=[]
for i in range(len(SW_del)):
    SW_del2.append(list(SW_del[i]))
```

```python
# 확인

print(SW_del2[2])
print(data_st[2])
```

```python
# data 컬럼에 추가

data['n_pos_t']=SW_del2
```

```python
# 확인

data['n_pos_t']
```

# content 불용어 제거를 위한 전처리

```python
# pos_c 리스트안의 문자형 분리 & 특수문자 제거

data_sp_c=[]
for i in range(len(data)):
    data_sp_c.append(data['pos_c'][i].split(','))
    
print(data_sp_c[0])
```

```python
data_st_c=[]
for i in range(len(data)):
    data_st2_c=[]
    for j in data_sp_c[i]:
        data_st2_c.append(j.lstrip('[').rstrip('"]').strip(" ' "))
    data_st_c.append(data_st2_c)
    
data_st_c[0][1]
```

# content 불용어 제거

```python
# 불용어 제거
SW_del_c=[]
for i in data_st_c:
    b2=set(i)
    SW_del_c.append(b2-a)
```

```python
# 확인

SW_del_c[:2]
```

```python
# set 변환이 된 데이터를 다시 리스트로 변환

SW_del2_c=[]
for i in range(len(SW_del_c)):
    SW_del2_c.append(list(SW_del_c[i]))
```

```python
# 확인

print(SW_del2_c[0])
print(data_st_c[0])
```

```python
# data 컬럼에 추가

data['n_pos_c']=SW_del2_c
```

```python
# 확인

data['n_pos_c']
```
