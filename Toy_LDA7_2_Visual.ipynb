```python
import matplotlib.pyplot as plt
data.plot()
plt.title("D1f와 Do")
plt.xlabel("시간")
plt.ylabel("Data")
plt.show()
```

# 축의 범위가 달라서 df1이 안보임 >  각각의 축을 따로 잡자

```python
df=data
```

```python
ax0=df.index
y1=df['Df']
y2=df['topic0']
```

```python
fig, ax0 = plt.subplots(figsize=(15,8))
ax1 = ax0.twinx() #복수의 y축을 가진 plot 그림 
ax0.set_title("Df1&D0")
ax0.plot(y1, 'r-', label="Df1")
ax0.set_ylabel("Df1")
ax0.grid(False)
ax1.plot(y2, 'g:', label="D0")
ax1.set_ylabel("D0")
ax1.grid(False)
ax0.set_xlabel("시간")
plt.figure().show() #사이즈 키워도 그림이 안커진다,, 
```
