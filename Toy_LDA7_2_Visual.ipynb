```python
import pandas as pd 
import matplotlib.pyplot as plt

# 파일 로드
D1=pd.read_excel('Df_topic0.xlsx')
D0=pd.read_excel('D0_topic0.xlsx')
```

```python
# 확인

D1.head()
```

```python
# 확인

D0.head()
```

```python
# merge

data=pd.merge(D0,D1)
```

```python
# 확인

data.head()
```

```python
# 확인

data.info()
```

```python
# plot 그리려면 x축이 될 기간을 index로 설정해야함

data.set_index(data['ymd'],inplace=True)
```

```python
# 컬럼 삭제


del data['ymd']
```

```python
# 확인

data.head()
```

```python
# 시각화

data.plot()
plt.title("D1f와 Do")
plt.xlabel("시간")
plt.ylabel("Data")
plt.show()
```

# 축의 범위가 달라서 df1이 안보임 >  각각의 축을 따로 잡음

```python
# df 변수 생성

df=data
```

```python
# y축 설정

ax0=df.index
y1=df['Df']
y2=df['topic0']
```

```python
# 시각화

fig, ax0 = plt.subplots(figsize=(15,8))
ax1 = ax0.twinx() #복수의 y축을 가진 plot 그림 
ax0.set_title("Df1&D0")
ax0.plot(y1, 'r-', label="Df1")
ax0.set_ylabel("Df1")
ax0.grid(False)
ax1.plot(y2, 'g:', label="D0")
ax1.set_ylabel("D0")
ax1.grid(False)
ax0.set_xlabel("시간")
plt.figure().show()
```
